### Classe van Emde Boas e suas Funções

```python
class veb:
    def getC(self, x):
        return int(math.floor(x / math.sqrt(self.u)))
    def getI(self, x):
        return int((x % math.ceil(math.sqrt(self.u))))
    def getX(self, c, i):
        return int((c * math.floor(math.sqrt(self.u))) + i)
    def __init__(self, u):
        if(u < 0):
            raise Exception("u cannot be less than 0 --- u = " + str(u));
        self.u = 2;
        while(self.u < u):
            self.u *= self.u
        self.min = None
        self.max = None
        if(u > 2):
            self.clusters = []
            for i in range(self.getC(self.u)):
                self.clusters.append(None)
        self.sumario = None
    def buscar(self, x):
        if(x == self.min or x == self.max):
            return True
        elif(self.u <= 2):
            return False
        else:
            cluster = self.clusters[self.getC(x)]
            if cluster != None:
                return cluster.buscar(self.getI(x))
            else:
                return False
    def incluirUnicoElemento(self, x):
        self.min = x
        self.max = x
    def incluir(self, x):
        if(not self.buscar(x)):
            if(self.min == None):
                self.incluirUnicoElemento(x)
                return True
            else:
                if(x < self.min):
                    temporario = self.min
                    self.min = x
                    x = temporario
                if(self.u > 2):
                    c = self.getC(x)
                    if(self.clusters[c] == None):
                        self.clusters[c] = veb(self.getC(self.u))
                    if(self.sumario == None):
                        self.sumario = veb(self.getC(self.u))
                    if(self.clusters[c].min == None):
                        self.sumario.incluir(c)
                        self.clusters[c].incluirUnicoElemento(self.getI(x))
                    else:
                        self.clusters[c].incluir(self.getI(x))
                if(x > self.max):
                    self.max = x
                return True
        else:
            return False
    def remover(self, x):
        if(self.buscar(x)):
            if(x == self.min):
                if(self.sumario == None):
                    c = None
                else:
                    c = self.sumario.min
                if(c == None):
                    self.min = None
                    return True
                self.min = self.getX(self.getC(x), self.clusters[c].min)
                x = self.min
            c = self.getC(x)
            self.clusters[c].remover(self.getI(x))
            if(self.clusters[c].min == None):
                self.sumario.remover(c)
            if(self.sumario.min == None):
                self.max = self.min
            else:
                cAux = self.sumario.max
                self.max = self.getX(cAux, self.clusters[cAux].max)
            return True
        else:
            return False
    def sucessor(self, x):
        if(self.u <= 2):
            if(x == 0 and self.max == 1):
                return 1
            else:
                return None
        elif(self.min != None and x < self.min):
            return self.min
        else:
            c = self.getC(x)
            i = self.getI(x)
            maxClusterC = None
            cluster = self.clusters[c]
            if(cluster != None):
                maxClusterC = cluster.max
            if(maxClusterC != None and i < maxClusterC):
                clusterSucc = cluster.successor(i)
                return self.getX(c, clusterSucc)
            else:
                clusterSucc = None
                if(self.sumario != None):
                    clusterSucc = self.sumario.sucessor(c)
                if(clusterSucc == None):
                    return None
                else:
                    cluster2 = self.clusters[clusterSucc]
                    minCluster = 0
                    if(cluster2 != None):
                        minCluster = cluster2.min
                    return self.getX(clusterSucc, minCluster)
    def predecessor(self, x):
        if self.u <= 2:
            if x == 1 and self.min == 0:
                return 0
            else:
                return None
        elif self.max != None and x > self.max:
            return self.max
        else:
            c = self.getC(x)
            i = self.getI(x)
            minClusterC = None
            cluster = self.clusters[c]
            if cluster != None:
                minClusterC = cluster.min
            if minClusterC != None and i > minClusterC:
                clusterPred = cluster.predecessor(i)
                if clusterPred == None: 
                    clusterPred = 0
                return self.getX(c, clusterPred)
            else:
                clusterPred = None
                if self.sumario != None:
                    clusterPred = self.sumario.predecessor(c)
                if clusterPred == None:
                    if self.min != None and x > self.min:
                        return self.min
                    else:
                        return None
                else:
                    cluster2 = self.clusters[clusterPred]
                    maxCluster = 0
                    if cluster2 != None:
                        maxCluster = cluster2.max
                    return self.index(clusterPred, maxCluster)
```

### Execução

```python
logs = []
```

```python
def executar():
    v = veb(64)
    entrada = open('entrada.txt','r')
    for comando in entrada:
        print(comando)
        com, valor = comando.split(':')
        if(com == 'INC'):
            v.incluir(int(valor))
            logs.append('INC:'+str(int(valor))+'\n\n')
        elif(com == 'REM'):
            v.remover(int(valor))
            logs.append('REM:'+str(int(valor))+'\n\n')
        elif(com == 'SUC'):
            suc = v.sucessor(int(valor))
            logs.append('SUC:'+str(int(valor))+'\n'+str(suc)+'\n\n')
        elif(com == 'PRE'):
            pred = v.predecessor(int(valor))
            logs.append('PRE:'+str(int(valor))+'\n'+str(pred)+'\n\n')
    entrada.close()
    saida = open('saida.txt', 'w')
    saida.writelines(logs)
    saida.close()
```

```python
executar()
```
